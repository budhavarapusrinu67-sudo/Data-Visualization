```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix, parallel_coordinates

# Load dataset (upload CSV to Colab and update filename)
df = pd.read_csv("/content/clean_dataset.csv")
print(df.columns)
```

```python
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

num_cols = ['Age', 'Debt', 'YearsEmployed', 'Income', 'CreditScore']

scatter_matrix(df[num_cols], figsize=(10, 10), diagonal='hist')
plt.suptitle("Scatterplot Matrix of Numerical Variables", y=1.02)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
from pandas.plotting import parallel_coordinates

features = ['Age', 'Debt', 'YearsEmployed', 'Income', 'CreditScore']

scaler = MinMaxScaler()
df_scaled = df.copy()
df_scaled[features] = scaler.fit_transform(df[features])

parallel_coordinates(df_scaled[['Approved'] + features], 'Approved')
plt.title("Parallel Coordinates Plot")
plt.show()
```

```python
plt.plot(df['Age'], df['Income'])
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Line Graph: Age vs Income")
plt.show()
```

```python
pd.crosstab(df['Gender'], df['Approved']).plot(
    kind='bar', stacked=True
)

plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Stacked Bar Chart: Gender vs Approval Status")
plt.show()
```
