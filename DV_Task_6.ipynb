```python
!pip install networkx
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
links = pd.read_csv('/content/stack_network_links (1).csv')
nodes = pd.read_csv('/content/stack_network_nodes.csv')
links.head()
```

```python
filtered_links = links[links['value'] > 20]
G = nx.from_pandas_edgelist(filtered_links, 'source', 'target', create_using=nx.Graph())
largest_cc = max(nx.connected_components(G), key=len)

G_largest = G.subgraph(largest_cc)

plt.figure(figsize=(14,12))
pos = nx.spring_layout(G_largest, k=0.3)

nx.draw(G_largest, pos,
        node_size=600,
        node_color="skyblue",
        with_labels=True,
        font_size=8)

plt.title("Largest Connected Component - Force Layout")
plt.axis('off')
plt.show()
```
