```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/synthetic_salary_dataset.csv')
print(df.columns.tolist())
```

```python
plt.figure()
sns.countplot(x='EducationLevel', data=df)
plt.title('Bar Chart of Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.show()
```

```python
industry_counts = df['Industry'].value_counts()

plt.figure()
plt.pie(industry_counts, labels=industry_counts.index, autopct='%1.1f%%')
plt.title('Pie Chart of Industry Distribution')
plt.show()
```

```python
plt.figure()
plt.scatter(df.index, df['Salary(INR)'])
plt.title('Scatter Plot of Salary')
plt.xlabel('Employee Index')
plt.ylabel('Salary (INR)')
plt.show()
```

```python
plt.figure()
plt.plot(df['Salary(INR)'])
plt.title('Line Plot of Salary')
plt.xlabel('Employee Index')
plt.ylabel('Salary (INR)')
plt.show()
```

```python
plt.figure()
sns.stripplot(x='EducationLevel', y='Salary(INR)', data=df)
plt.title('Strip Plot of Salary by Education Level')
plt.show()
```

```python
plt.figure()
sns.swarmplot(x='EducationLevel', y='Salary(INR)', data=df)
plt.title('Swarm Plot of Salary by Education Level')
plt.show()
```

```python
plt.figure()
plt.hist(df['Salary(INR)'], bins=10)
plt.title('Histogram of Salary')
plt.xlabel('Salary (INR)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure()
sns.kdeplot(df['Salary(INR)'], fill=True)
plt.title('Density Plot of Salary')
plt.xlabel('Salary (INR)')
plt.show()
```

```python
plt.figure()
sns.rugplot(df['Salary(INR)'])
plt.title('Rug Plot of Salary')
plt.xlabel('Salary (INR)')
plt.show()
```
