```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Define the text data for the word cloud
text_data = "This is an example text to demonstrate a word cloud. Word clouds are a great way to visualize text data. They show the most frequent words in a prominent way."

# Generate word cloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white"
).generate(text_data)

# Display word cloud
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud - Text Network Analysis")
plt.show()
```
