```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
df = pd.read_csv("/content/StudentsPerformance.csv")
df.head()
```

```python
categorical_columns = df.select_dtypes(include='object').columns
continuous_columns = df.select_dtypes(include='number').columns

print("Categorical Columns:")
print(categorical_columns)

print("\nContinuous Columns:")
print(continuous_columns)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
ct = pd.crosstab(df['gender'], df['test preparation course'])
ct.plot(kind='bar', stacked=True)
plt.title("Stacked Bar Chart: Gender vs Test Preparation Course")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
ct.plot(kind='bar')
plt.title("Grouped Bar Chart: Gender vs Test Preparation Course")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
ct_percent = ct.div(ct.sum(axis=1), axis=0)

ct_percent.plot(kind='bar', stacked=True)
plt.title("Segmented Bar Chart (Percentage): Gender vs Test Preparation Course")
plt.xlabel("Gender")
plt.ylabel("Proportion")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot with regression (fit) line
sns.regplot(
    x='math score',
    y='reading score',
    data=df,
    scatter_kws={'alpha': 0.6},
    line_kws={'color': 'red'}
)

plt.title('Scatter Plot with Fit Line: Math Score vs Reading Score')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.show()
```

```python
sns.barplot(x='gender', y='math score', data=df)
plt.title("Average Math Score by Gender")
plt.show()
```

```python
sns.kdeplot(data=df, x='math score', hue='gender', fill=True)
plt.title("Grouped Kernel Density Plot: Math Score by Gender")
plt.show()
```

```python
sns.boxplot(x='gender', y='math score', data=df)
plt.title("Box Plot: Math Score by Gender")
plt.show()
```

```python
sns.violinplot(x='gender', y='math score', data=df)
plt.title("Violin Plot: Math Score by Gender")
plt.show()

```

```python
sns.kdeplot(data=df, x='math score', hue='gender', multiple="stack", fill=True)
plt.title("Ridgeline Plot: Math Score Distribution by Gender")
plt.show()
```

```python
sns.swarmplot(x='gender', y='math score', data=df)
plt.title("Beeswarm Plot: Math Score by Gender")
plt.show()
```
