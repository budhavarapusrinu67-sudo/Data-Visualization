```python
import plotly.express as px
import pandas as pd
import plotly.express as px
df = pd.read_csv('/content/sales data.csv')
df.columns
```

```python
fig = px.treemap(
    df,
    path=['Region', 'Category', 'Sub-Category'],
    values='Sales',
    title='TreeMap of Sales by Region, Category and Sub-Category'
)
fig.show()
```

```python
fig2 = px.sunburst(
    df,
    path=['Region', 'Category', 'Sub-Category'],
    values='Sales',
    title='Sunburst Chart of Sales by Region, Category and Sub-Category'
)

fig2.show()
```
