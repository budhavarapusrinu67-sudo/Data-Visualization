```python
import networkx as nx
import matplotlib.pyplot as plt
import re
```

```python
text = """
Data visualization is the graphical representation of data and information.
It helps people understand complex data through visual elements such as charts, graphs, and maps.
Data visualization makes it easier to identify patterns, trends, and relationships in datasets.
Tools like Python, Matplotlib, Tableau, and Power BI are commonly used for data visualization.

"""

words = re.findall(r'\b\w+\b', text.lower())

G = nx.DiGraph()

for i in range(len(words)-1):
    G.add_edge(words[i], words[i+1])

pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels=True, node_size=1000, font_size=8)

plt.title("Word Tree Visualization")
plt.show()
```
